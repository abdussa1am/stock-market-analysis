```python

data = cs.loc[:2000,['Unnamed: 0','HBL_open','HBL_close']]
data = data.rename(columns={'Unnamed: 0': 'Date'})
#data = data.set_index('Date')
data.Date = pd.to_datetime(data.Date)
data.set_index('Date', inplace=True)
#data = data['2018-05-04':'2010-04-16':-1]
data = data.sort_index(axis=0, ascending=True)
data
```

```python
import math
import numpy as np
new = data.filter(['HBL_open'])
dt = new.values
print(len(dt))
training_data_len = math.ceil(len(new) * 0.8)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(new)
print(scaled_data)
```

```python
train_data = scaled_data[0:training_data_len , :]
len(train_data)
```

```python
x_train = []
y_train = []
for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
    if i<=61:
        print(x_train)
        print(y_train)
print("\n")
print(len(x_train[1500]))
```

```python
x_train , y_train = np.array(x_train),np.array(y_train)
x_train.shape
```

```python
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras import models
model = models.Sequential()
model.add(LSTM(50, return_sequences = True ,input_shape=(x_train.shape[1],1)))
model.add(LSTM(50, return_sequences = False))
model.add(Dense(50))
model.add(Dense(1))
 
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=1 ,epochs = 3)
```

```python
test_data = scaled_data[training_data_len-60: , :]
x_test = []
y_test = dt[training_data_len:, :]
for i in range(60 , len(test_data)):
    x_test.append(test_data[i-60:i, :])
len(x_test)
```

```python
x_test=np.array(x_test)
```

```python
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```python
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```python
train = new[:training_data_len]
valid = new[training_data_len:]
valid['predictions'] = predictions
plt.figure(figsize=(14,5))
plt.title("model")
plt.xlabel("date",fontsize=18)
plt.ylabel("open", fontsize=18)
plt.plot(train["HBL_open"])
plt.plot(valid[['HBL_open','predictions']])
plt.legend(['train' ,'val','predictions'], loc = 'lower right')
plt.show()
```
